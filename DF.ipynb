```python
import os
import pandas as pd
```

```python
root_dir = '/content/drive/MyDrive/Film pano'
```

```python
!ls '/content/drive/MyDrive/Film pano'
```

```python
file_paths = []
genders = []
categories = []
ages = []
```

```python
exclude_files = {'FILM NUMBER.gsheet', 'Detail pic.jpg'}
```

```python
# Walk through the directory structure
for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file not in exclude_files:
            full_path = os.path.join(root, file)
            file_paths.append(full_path)

            # Extract gender from path and map it
            if 'F/' in full_path:
                genders.append('Female')
            elif 'M/' in full_path:
                genders.append('Male')
            else:
                genders.append('Unknown')

            # Extract category from path
            if 'Both-' in full_path:
                categories.append('Both')
            elif 'Lt-' in full_path:
                categories.append('Lt')
            elif 'Rt-' in full_path:
                categories.append('Rt')
            else:
                categories.append('Unknown')

            # Extract age from path
            age_str = next((part for part in full_path.split('/') if 'Y' in part), None)
            if age_str:
                age = age_str.split('Y')[0]
                ages.append(int(age))
            else:
                ages.append('Unknown')
```

```python
df = pd.DataFrame({'File Path': file_paths, 'Gender': genders, 'Category': categories, 'Age': ages})
```

```python
df = df.sort_values(by='Age', key=lambda x: pd.to_numeric(x, errors='coerce'))
```
