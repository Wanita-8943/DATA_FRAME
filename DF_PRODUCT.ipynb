```python
import os
import pandas as pd

# Define the root directory (the folder you provided)
root_dir = '/content/drive/MyDrive/PRODUCT'  # Update this with the correct path

# Create a list to store file paths
file_paths = []

# Walk through the directory structure
for root, dirs, files in os.walk(root_dir):
    for file in files:
        # Create the full path
        full_path = os.path.join(root, file)
        # Append the full path to the list
        file_paths.append(full_path)

# Convert the list to a DataFrame
df = pd.DataFrame(file_paths, columns=['Path'])


def extract_product(path):
    # Split the path and find the product part
    parts = path.split('/')
    for part in parts:
        if part.startswith(('OISHI_', 'Fuji_')):
            return part
    return None


# Function to determine if the file is in Train or Validation folder
def extract_train_validation(path):
    if 'Train' in path:
        return 'Train'
    elif 'Validation' in path:
        return 'Validation'
    return None

df['Product'] = df['Path'].apply(extract_product)
df['TrainAndValidation'] = df['Path'].apply(extract_train_validation)
```
